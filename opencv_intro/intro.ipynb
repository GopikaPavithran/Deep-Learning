```python
#To resize image and window


# Get the current window size
width, height = img.shape[1], img.shape[0]

# Calculate the new width and height based on a desired scaling factor (e.g., 0.5 for half size)
scaling_factor = 0.7
new_width = int(width * scaling_factor)
new_height = int(height * scaling_factor)

# Resize the image
resized_img = cv2.resize(img, (new_width, new_height))
```

```python
img_gray = cv2.imread('A Cute Cat Photo _ JPG Free Download - Pikbest.jpeg',cv2.IMREAD_GRAYSCALE)
img_gray
```

```python
#array([[250, 250, 250, ..., 243, 243, 243],
#        [250, 250, 250, ..., 243, 243, 243],
#        [250, 250, 250, ..., 243, 243, 243],
#        ...,
#        [152, 152, 153, ..., 203, 211, 214],
#        [152, 152, 152, ..., 205, 213, 219],
#        [152, 152, 151, ..., 199, 212, 231]], dtype=uint8)

# 2D array , each number represents a pixel
```

```python
img_gray.shape
```

```python
w,h=img_gray.shape[1],img_gray.shape[0]
scaling_factor=0.7
new_w=int(scaling_factor*w)
new_h=int(scaling_factor*h)
img_gray=cv2.resize(img_gray,(new_w,new_h))
```

```python
cv2.imshow("Cat Gray",img_gray)
cv2.waitKey()
cv2.destroyAllWindows()
```

```python
thresh,img_binary=cv2.threshold(img_gray,170,255,cv2.THRESH_BINARY)   # threshold() -  to convert a grayscale image into a binary image
cv2.imshow('Cat Binary',img_binary)                                   # 170 -  determines the cutoff point for separating pixels into 
cv2.waitKey()                                                         #         black and white. Pixels with values(from gray sacle image)
cv2.destroyAllWindows()                                               #         greater than or equal to 125 will be set to 255 (white), 
                                                                      #         and pixels with values less than 125 will be set to 0 (black).
```

```python
thresh
```

```python
thresh,img_binary=cv2.threshold(img_gray,125,255,cv2.THRESH_OTSU)    #THRESH_OTSU - to automatically determine the optimal threshold value
cv2.imshow('Cat Binary',img_binary)
cv2.waitKey()
cv2.destroyAllWindows()
```

```python
#matplotlib

import matplotlib.pyplot as plt
plt.imshow(img)
```

```python
# When we load an image(say an rgb image) using cv2.imread(), it loads the image in BGR format by default.and when we 
# need to diaplay the image using cv2.imshow() it will automatically show the image in correct format(rgb)
# but when we use matplotlib to diaplay image it will not convert the bgr image, we need to convert it manually.
```

```python
conv_img=cv2.cvtColor(img,cv2.COLOR_BGR2RGB)
plt.imshow(conv_img)
```

```python
# X axis -> values increses from left to right
# Y axis -> values increases from top to bottom
```

```python
#drawing rectangle - we need to define 2 points to draw a rectangle (top left and bottom right)

conv_img=cv2.cvtColor(img,cv2.COLOR_BGR2RGB)   # include the conversion ineach step of the drawing process
                                               # to avoid overlapping multiple drawings on a single image
cv2.rectangle(conv_img,pt1=(150,90),pt2=(530,610),color=(32,0,224),thickness=3)
plt.imshow(conv_img)                           # negative value for thickness will fill the rectangle
```

```python
#Circle

conv_img=cv2.cvtColor(img,cv2.COLOR_BGR2RGB)
cv2.circle(conv_img,center=(350,350),radius=290,color=(34,0,0),thickness=3)
plt.imshow(conv_img)
```

```python
#putText

conv_img=cv2.cvtColor(img,cv2.COLOR_BGR2RGB)
cv2.putText(conv_img,text="Kitty",org=(300,50),fontFace=cv2.FONT_HERSHEY_COMPLEX_SMALL,fontScale=2,color=(3,8,6),thickness=3)
plt.imshow(conv_img)
```
